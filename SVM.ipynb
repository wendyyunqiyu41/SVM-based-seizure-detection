```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal
sf = 256
# Define window length (4 seconds)
win = 4 * sf
freqs, psd = signal.welch(hour1FP1F3, sf, nperseg=win)

# Plot the power spectrum
sns.set(font_scale=1.2, style='white')
plt.figure(figsize=(8, 4))
plt.plot(freqs, psd, color='k', lw=2)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power spectral density (V^2 / Hz)')
plt.ylim([0, psd.max() * 1.1])
plt.title("Welch's periodogram")
plt.xlim([0, freqs.max()])
sns.despine()
```

```python
# Define delta lower and upper limits
low, high = 0.5, 4

# Find intersecting values in frequency vector
idx_delta = np.logical_and(freqs >= low, freqs <= high)

# Plot the power spectral density and fill the delta area
plt.figure(figsize=(7, 4))
plt.plot(freqs, psd, lw=2, color='k')
plt.fill_between(freqs, psd, where=idx_delta, color='skyblue')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power spectral density (uV^2 / Hz)')
plt.xlim([0, 10])
plt.ylim([0, psd.max() * 1.1])
plt.title("Welch's periodogram")
sns.despine()
```

```python
from scipy.integrate import simps

# Frequency resolution
freq_res = freqs[1] - freqs[0]  # = 1 / 4 = 0.25
print(freq_res,freqs[1],freqs[0])

# Compute the absolute power by approximating the area under the curve
delta_power = simps(psd[idx_delta], dx=freq_res)
print('Absolute delta power: %.3f uV^2' % delta_power)

# Relative delta power (expressed as a percentage of total power)
total_power = simps(psd, dx=freq_res)
delta_rel_power = delta_power / total_power
print('Relative delta power: %.3f' % delta_rel_power)
```

```python
# allhoursallchannels = {'FP1F7':[],'F7T7':[],'T7P7':[],'P7O1':[],
#                   'FP1F3':[],'F3C3':[],'C3P3':[],'P3O1':[],
#                   'FP2F4':[],'F4C4':[],'C4P4':[],'P4O2':[],
#                   'FP2F8':[],'F8T8':[],'T8P8':[],'P8O2':[],
#                   'FZCZ':[],'CZPZ':[],'P7T7':[],'T7FT9':[],
#                   'FT9FT10':[],'FT10T8':[]}
# channels = ['FP1F7','F7T7','T7P7','P7O1',
#                   'FP1F3','F3C3','C3P3','P3O1',
#                   'FP2F4','F4C4','C4P4','P4O2',
#                   'FP2F8','F8T8','T8P8','P8O2',
#                   'FZCZ','CZPZ','P7T7','T7FT9',
#                   'FT9FT10','FT10T8']

allhoursallchannels = {
                        'F3C3':[], 'C3P3':[], 'F4C4':[], 
                        'C4P4':[], 'CZPZ':[], 'P7T7':[]
                        }
channels = [
            
            'F3C3', 'C3P3', 'F4C4', 
            'C4P4', 'CZPZ', 'P7T7'
            ]

# allhoursallchannels = {'F3-C3':[],'C3-P3':[],'F4-C4':[],'C4-P4':[],'CZ-PZ':[],'P7-T7':[]}
# channels = ['F3-C3','C3-P3','F4-C4','C4-P4','CZ-PZ','P7-T7']
# channelcolumn = [9,10,12,13,20,21]
datafiles = ['data/chb24/chb24_01.txt',
            'data/chb24/chb24_02.txt',
            'data/chb24/chb24_03.txt',
            'data/chb24/chb24_04.txt',
            'data/chb24/chb24_05.txt',
            'data/chb24/chb24_06.txt',
            'data/chb24/chb24_07.txt',
            'data/chb24/chb24_08.txt',
            'data/chb24/chb24_09.txt',
            'data/chb24/chb24_10.txt',
            'data/chb24/chb24_11.txt',
            'data/chb24/chb24_12.txt',
            'data/chb24/chb24_13.txt',
            'data/chb24/chb24_14.txt',
            'data/chb24/chb24_15.txt',
            'data/chb24/chb24_16.txt',
            'data/chb24/chb24_17.txt',
            'data/chb24/chb24_18.txt',
            'data/chb24/chb24_19.txt',
            'data/chb24/chb24_20.txt',
            'data/chb24/chb24_21.txt',
            'data/chb24/chb24_22.txt']


for i in range(len(datafiles)): 
    print(i)
    with open(datafiles[i]) as hour:
        next(hour)
        for line in hour:
            for j in range(len(channels)):
                allhoursallchannels[channels[j]].append(float(line.strip().split('\t')[j+8]))

for i in range(len(channels)):
    print(allhoursallchannels[channels[i]][-10:])
```

```python
from scipy import signal
filteredallhoursallchannels = {
                        'F3C3':[], 'C3P3':[], 'F4C4':[], 
                        'C4P4':[], 'CZPZ':[], 'P7T7':[]
                        
                        }
fs = 256.0  # Sample frequency (Hz)
f0 = 60.0  # Frequency to be removed from signal (Hz)
Q = 30.0  # Quality factor
# Design notch filter
b, a = signal.iirnotch(f0, Q, fs)
for i in range(len(channels)):
    print(channels[i])
    filteredallhoursallchannels[channels[i]] = signal.lfilter(b, a, allhoursallchannels[channels[i]])
for i in range(len(channels)):
    print(filteredallhoursallchannels[channels[i]][-10:])
```

```python
# Segment & label data of preictal(P) & interictal(N) class
preS = []
preE = []
nonS = []
nonE = []
with open('preictal_intervals.txt') as table:
    next(table)
    for line in table:
        preS.append(float(line.strip().split('\t')[0]))
        preE.append(float(line.strip().split('\t')[1]))
        
with open('interictal_intervals.txt') as table:
    next(table)
    for line in table:
        nonS.append(float(line.strip().split('\t')[0]))
        nonE.append(float(line.strip().split('\t')[1]))

preTime = np.vstack((np.array(preS),np.array(preE)))
nonTime = np.vstack((np.array(nonS),np.array(nonE)))
pre_pos = preTime.astype(int) * 256
non_pos = nonTime.astype(int) * 256
print(preTime)
print(nonTime)
print(pre_pos)
print(non_pos)
```

```python
# Find start & end index of each inverval for spectral power calculation
winSize = 20 # window size in sec
pre_win = np.floor_divide((preTime[1,:]-preTime[0,:]),winSize/2).astype(int) - 1 
non_win = np.floor_divide((nonTime[1,:]-nonTime[0,:]),winSize/2).astype(int) - 1
batch = int(pre_win[1])

# num overlapping windows for each interval
pre_size = int(np.sum(pre_win))
non_size = int(np.sum(non_win))
print('One batch of data in 11-fold cross validation has size:', batch)
print('Preictal total number of windows(samples)',pre_size)
print('Interictal total number of windows(samples):',non_size)

print('# 20s sliding windows in each batch')
print(pre_win)
print(non_win)

# Update ending position of each batch for indexing purpose
pre_pos[1,:] = pre_pos[0,:] + (pre_win + 1)*10*256
non_pos[1,:] = non_pos[0,:] + (non_win + 1)*10*256
print('\nPreictal data batch star&end postitions(index):\n', pre_pos)
print('Interictal start&end postitions(index): \n', non_pos)
```

```python
# Calculate spectral power of each windows & save feature values
frequencybandsname = ['delta','theta','alpha','beta','gamma1','gamma2','total']
frequencybandslow = [0.5,4,8,13,30,63,0.5]
frequencybandshigh = [4,8,13,30,57,128,128]
window = [4*256,0.5*256,0.25*256,0.153846*256,0.066667*256,0.031746*256,4*256]
# channels: {'FP1F3':[],'F3C3':[],'C3P3':[],'FP2F4':[],'F4C4':[],'C4P4':[]}

# Preictal class
preictalfeature = np.empty([pre_size,len(channels)*len(window)])

preID = 0
for interval in range(len(pre_win)): # iter thru 11 intervals
    print(interval)
    for j in range(pre_win[interval]): # iter thru 78 windows in a interval
        start = pre_pos[0,interval] + j * int(winSize/2) * sf
        end = start + winSize * sf
        for i in range(len(channels)): # iter thru 6 channels
            for k in range(len(window)): # iter thru 7 frequency bands
                freqs, psd = signal.welch(filteredallhoursallchannels[channels[i]][start:end], sf, nperseg=window[k])
                idx_band = np.logical_and(freqs >= frequencybandslow[k], freqs <= frequencybandshigh[k])
#                 print(interval,j,i,k,freqs[1],freqs[0],freqs[1]-freqs[0])
                freq_res = freqs[1] - freqs[0]  # = 1 / 4 = 0.25
                bandpower = simps(psd[idx_band], dx=freq_res)
                totalpower = simps(psd, dx=freq_res)
                relpower = bandpower / totalpower
                preictalfeature[preID,i*len(window) + k] = relpower
        preID += 1

# Interictal class
interictalfeature = np.empty([non_size,len(channels)*len(window)])

nonID = 0
for interval in range(len(non_win)): # iter thru 5 intervals
    print(interval)
    for j in range(non_win[interval]): # iter thru N windows in a interval
        start = pre_pos[0,interval] + j * int(winSize/2) * sf
        end = start + winSize * sf
        for i in range(len(channels)): # iter thru 6 channels
            for k in range(len(window)): # iter thru 7 frequency bands
                freqs, psd = signal.welch(allhoursallchannels[channels[i]][start:end], sf, nperseg=window[k])
                idx_band = np.logical_and(freqs >= frequencybandslow[k], freqs <= frequencybandshigh[k])
#                 print(interval,j,i,k,freqs[1],freqs[0],freqs[1]-freqs[0])
                freq_res = freqs[1] - freqs[0]  # = 1 / 4 = 0.25
                bandpower = simps(psd[idx_band], dx=freq_res)
                totalpower = simps(psd, dx=freq_res)
                relpower = bandpower / totalpower
                interictalfeature[nonID,i*len(window) + k] = relpower
        nonID += 1
```

```python
from sklearn import svm
# Randomly draw data & group for 11-fold cross validation
fold = 11
dim = 7*6
samples = np.arange(non_size)

pre_dataID = np.arange(pre_size).reshape([11,batch])
non_dataID = np.random.choice(samples,size=pre_size,replace=False).reshape([11,batch])

# 11-fold cross validation using SVM
gamma = np.linspace(0.0001,0.05,num=30)
percent_mean = np.empty(len(gamma))
for g in range(len(gamma)):
    print(g)
    result = np.empty((batch*2,fold))
    for f in range(fold):
        trainX_pre = preictalfeature[np.delete(pre_dataID,f,0),:].reshape((batch*(fold-1),dim))
        trainX_non = interictalfeature[np.delete(non_dataID,f,0),:].reshape((batch*(fold-1),dim))
        trainX = np.vstack((trainX_pre, trainX_non))
        trainY = [1] * (pre_size-batch) + [0] * (pre_size-batch)

        clf = svm.SVC(gamma=gamma[g])
        clf.fit(trainX, trainY)
        testX = np.vstack((preictalfeature[pre_dataID[f,:],:], interictalfeature[non_dataID[f,:],:]))
        testY = np.vstack((np.ones((batch, fold)), np.zeros((batch,fold))))
        result[:,f] = clf.predict(testX)

    correct = np.count_nonzero(np.equal(result,testY),axis=0)
    percent = correct/len(testX)
    percent_mean[g] = sum(percent)/fold*100
    print('gamma = ', gamma[g],'Classification accuracy is:', percent_mean[g],'%')
```

```python
plt.figure(figsize=(8, 4))
plt.plot(gamma, percent_mean, lw=2, color='k')
#plt.fill_between(freqs, psd, where=idx_delta, color='skyblue')
plt.xlabel('Gamma')
plt.ylabel('Percent of Correct Classification (%)')
plt.xlim([0, gamma.max()])
plt.ylim([percent_mean.min()-0.5, percent_mean.max()])
plt.title("Classification Correct % VS Gamma")
sns.despine()
```

```python
from sklearn import datasets
#from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix

# Plot non-normalized confusion matrix
# titles_options = [("Confusion matrix, without normalization", None),
#                   ("Normalized confusion matrix", 'true')]
titles_options = [("Normalized confusion matrix", 'true')]
class_names = ['Preictal','Interictal']
confMat = np.zeros((2,2))

for f in range(fold):
    trainX_pre = preictalfeature[np.delete(pre_dataID,f,0),:].reshape((batch*(fold-1),dim))
    #trainX_non = interictalfeature[np.delete(non_dataID,f,0),:].reshape((batch*(fold-1),dim))
    trainX_non = interictalfeature[np.delete(non_dataID,f,0),:].reshape((batch*(fold-1),dim))
    trainX = np.vstack((trainX_pre, trainX_non))
    #trainY = [1] * (pre_size-batch) + [0] * (pre_size-batch)
    trainY = [1] * (pre_size-batch) + [0] * (batch*(fold-1))
    clf = svm.SVC(gamma='auto')
    clf.fit(trainX, trainY)
    testX = np.vstack((preictalfeature[pre_dataID[f,:],:], interictalfeature[non_dataID[f,:],:]))
    #testY = np.vstack((np.ones((batch, fold)), np.zeros((batch,fold))))
    testY = np.vstack((np.ones((batch, 1)), np.zeros((batch,1))))
    
    for title, normalize in titles_options:
        disp = plot_confusion_matrix(clf, testX, testY,
                                     display_labels=class_names,
                                     cmap=plt.cm.Blues,
                                     normalize=normalize)
        disp.ax_.set_title(title)
    confMat += np.vstack((disp.confusion_matrix[0], disp.confusion_matrix[1]))
plt.show()
print(np.true_divide(confMat,fold))
```
